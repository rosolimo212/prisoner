```python
logs = prs.make_game(
    player1=prs.random_bot, 
    player2=prs.random_bot, 
    max_distance=15
    )
```

```python
df = pd.DataFrame(logs)
```

```python
df.sample(3)
```

```python
df['game_id'] = 15
game_stat = df.groupby('game_id').agg(
    rounds = ('round', 'count'),
    player1_total_score = ('player1_round_score', 'sum'),
    player2_total_score = ('player2_round_score', 'sum'),
    player1_cooperates = ('player1', lambda x: (x == 1).sum()),
    player2_cooperates = ('player2', lambda x: (x == 1).sum()),
    player1_defects = ('player1', lambda x: (x == 0).sum()),
    player2_defects = ('player2', lambda x: (x == 0).sum()),
).reset_index()
game_stat['player1_round_score_per_round'] = game_stat['player1_total_score'] / game_stat['rounds']
game_stat['player2_round_score_per_round'] = game_stat['player2_total_score'] / game_stat['rounds']
game_stat['player1_cooperates_per_round'] = game_stat['player1_cooperates'] / game_stat['rounds']
game_stat['player2_cooperates_per_round'] = game_stat['player2_cooperates'] / game_stat['rounds']
game_stat['player1_defects_per_round'] = game_stat['player1_defects'] / game_stat['rounds']
game_stat['player2_defects_per_round'] = game_stat['player2_defects'] / game_stat['rounds']
game_stat
```
